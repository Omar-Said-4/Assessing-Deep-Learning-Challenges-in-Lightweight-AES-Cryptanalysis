```python
#import aes
import random
import csv
import os
from Crypto.Cipher import AES
from Crypto.Util import Counter
```

```python
def generate_random_128bit():
    return random.getrandbits(128)
```

```python
def int_to_hex_str(value):
    return hex(value)[2:].zfill(32)
```

```python
num_pairs = 10000000
output_file = 'aes_Quadruplets.csv'
```

```python
with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['plaintext', 'ciphertext', 'key', 'iv', 'keystream'])
    
    for i in range(0, num_pairs, 5):
        mk = generate_random_128bit()  # Master key
        iv = generate_random_128bit()  # Initial vector

        # Create a Counter object with the generated IV
        ctr = Counter.new(128, initial_value=int.from_bytes(iv.to_bytes(16, byteorder='big'), byteorder='big'))
        
        # Initialize AES cipher in CTR mode
        mk = mk.to_bytes(16, byteorder='big')
        cipher = AES.new(mk, AES.MODE_CTR, counter=ctr)
        
        for j in range(5):
            if i + j >= num_pairs:
                break
            
            pt = generate_random_128bit()  # Plaintext
            pt_bytes = pt.to_bytes(16, byteorder='big')

            # Encrypt the plaintext
            ct = cipher.encrypt(pt_bytes)

            keystream = int.from_bytes(ct, byteorder='big') ^ int.from_bytes(pt_bytes, byteorder='big')

            # Convert to hexadecimal strings for output
            pt_hex = int_to_hex_str(pt)
            ct_hex = int_to_hex_str(int.from_bytes(ct, byteorder='big'))
            mk_hex = int_to_hex_str(int.from_bytes(mk, byteorder='big'))
            iv_hex = int_to_hex_str(iv)
            keystream_hex = int_to_hex_str(keystream)


            # Write to the CSV file
            writer.writerow([pt_hex, ct_hex, mk_hex, iv_hex, keystream_hex])

            # Progress indicator (optional)
            if (i + j) % 100000 == 0:
                print(f'Processed {i + j} pairs')

print(f'Generated {num_pairs} AES triplets and saved to {output_file}')
```
