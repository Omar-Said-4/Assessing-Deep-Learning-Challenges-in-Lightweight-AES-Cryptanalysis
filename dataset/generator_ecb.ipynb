```python
from saes import SimplifiedAES as saes
import random
import csv
import os
from collections import defaultdict
```

```python
def generate_random_32bit():
    return random.getrandbits(32)
```

```python
def generate_random_16bit():
    return random.getrandbits(16)
```

```python
def int_to_hex_str(value):
    return hex(value)[2:].zfill(4)
```

```python
num_pairs = 1000000
output_file = '30k_pt_each_key_used_with_15_pt.csv'
```

```python
used_keys = set()
```

```python
def generate_unique_plaintext_values(count):
    # Use a set to ensure unique values
    unique_values = set()
    while len(unique_values) < count:
        unique_values.add(generate_random_16bit())
    return list(unique_values)
```

```python
max_plaintexts = 65536
pt_values = generate_unique_plaintext_values(max_plaintexts)
random.shuffle(pt_values)
print(pt_values)
```

```python
pts1=pt_values[:20001]
pts2=pt_values[20001:60000]
```

```python
keys = generate_unique_plaintext_values(65536)
random.shuffle(keys)
print(keys)
```

```python
keys1=keys[:30001]
keys2=keys[30001:50000]
```

```python
with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['plaintext', 'ciphertext', 'key'])
    for i in range(len(keys)):  # Generate 15 ciphertexts for each key
        key = keys[i]

        # Encrypt 20 plaintexts using the same key
        for j in range(15):  # Iterate over 15 plaintexts
            pt = pts2[(i + j)%20000]  # Get the 16-bit plaintext

            # Encrypt the 16-bit plaintext using SimplifiedAES
            cipher = saes(key)
            ct = cipher.encrypt(pt)  # Encrypt the 16-bit plaintext

            # Convert to hexadecimal strings for output
            pt_hex = int_to_hex_str(pt)
            ct_hex = int_to_hex_str(ct)
            key_hex = int_to_hex_str(key)

            # Write to the CSV file
            writer.writerow([pt_hex, ct_hex, key_hex])

        # Progress indicator (optional)
        # Progress indicator (optional)
        if (i ) % 100000 == 0:
            print(f'Processed {i} pairs')

print(f'Generated {num_pairs} SAES triplets and saved to {output_file}')
```
